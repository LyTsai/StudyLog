```python
import numpy as np
```

```python
operators = ['x', '/', '-', '+']
numbers = [3, 4, 5, 9]
value = 0
count = 0
```

```python
while value != 24 or count > 1000:
  count += 1
  np.random.shuffle(numbers)
  for i in range(0, len(numbers)):
    number = numbers[i]
    if i == 0:
        express = str(number)
        value = number
        continue
    operator = np.random.choice(operators)
    if operator == '+':
      value += number
    elif operator == '-':
      value -= number
    elif operator == 'x':
      value *= number
    else:
      value /= number

    express += operator + str(number)


print(express, value, count)
```
