```python
import numpy as np
```

```python
operators = ['x', '/', '-', '+']
numbers = [3, 4, 5, 9]
value = 0
count = 0
```

```python
while value != 24 | count < 1000:
  value = 0
  count += 1
  index = np.random.randint(0, 3)
  print(index)
  first_n = numbers[index]
  express = first_n
  for i in range(0, 3):
    number = numbers[i]
    operator = np.random.choice(operators)
    print(operator)
    if operator == '+':
      value += number
    elif operator == '-':
      value -= number
    elif operator == 'x':
      value *= number
    else:
      value /= number

    express += operator + str(number)


print(express)
```
